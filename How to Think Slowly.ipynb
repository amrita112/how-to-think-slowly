```python
# Load some packages - these contain pre-defined functions used to generate the plots
import numpy as np
import matplotlib.pyplot as plt
from os.path import sep

np.random.seed(0)
```

# Parameters

These are the numbers you can play with.

```python
# The total number of people, n, in the simulation. Larger numbers (>1000) make it easier to see the effect.
# You can use smaller numbers too, but the regression effect won't jump out as easily from the figures.
n = 5000 

# n_high is the number of people with the highest weights, selected to demonstrate regression to the mean 
# (the ones coloured in red in the figure). Technically, you can make this as large as 0.5n and still get
# regression--but keep it small, to about 0.01n, to see the effect clearly.
n_high = 50 

# The average or mean weight of the population, pop_mean_weight, is pretty arbitrary--you can set it to 
# whatever you want. I went with 75 (in kgs) based on a paper that I've linked below.
pop_mean_weight = 75 # Population mean weight

# This is the standard deviation of the distribution of weights (across the whole population). 
# The smaller this number, the more the weights will be bunched up around the mean weight. 
# I went with 1.2 kg based on the paper linked below.
pop_std_weight = 1.2 

# This is the standard deviation of the distribution of the weight of each individual, measured across days. 
# This determines how similar w1 and w2 are to each other (smaller std_weight --> more similar).
# As long as this number is > 0, you will see the regression effect. I went with 0.6 kg based on the article
# linked below.
std_weight = 0.6 
```

Here is the [article](https://www.scirp.org/journal/paperinformation?paperid=117728) that says average weight is 75 kg and the standard deviation is 1.2 kg. Two caveats:
1. This is based on American military personnel. It doesn't really matter because the point is to demonstrate the effect of selecting outliers in a population, which will show up in any population. This just happened to be the first paper I found with exact values of mean and standard deviation of weights in a population.
2. The paper shows that weights are actually not normally distributed--they are skewed to the right, i.e. there are more outliers on the heavy side than the light side--but I went with a normal distribution to keep things simple. Regression to the mean happens for any two correlated variables, the distribution doesn't matter.

Here is the [article](https://www.healthline.com/health/weight-fluctuation#other-factors) that talks about daily fluctuation in weight, about 2-3 kgs. In a normal distribution, 95% of the values fall between two standard deviations of the mean, so I took the standard deviation of each person's weight to be 2.5/4 = 0.6 kg (approx).

# Generate w1 and w2

```python
# Generate each person's mean weight. Note that in the real world weights don't strictly follow a normal 
# distribution, but we're doing that here for simplicity. 
mean_weight = np.random.normal(loc = pop_mean_weight, scale = pop_std_weight, size = n)

# Generate w1 and w2 for each person based on the mean weights we simulated above.
w1 = np.random.normal(loc = mean_weight, scale = std_weight)
w2 = np.random.normal(loc = mean_weight, scale = std_weight)

# Plot w1 and w2, and the X = Y line
plt.figure(figsize = [4, 4])
plt.scatter(w1, w2, color = 'w', marker = 'o', edgecolor = 'k', s = 20, alpha = 0.3)
plt.xlabel('w1')
plt.ylabel('w2')
min_weight = np.min([np.min(w1), np.min(w2)])
max_weight = np.max([np.max(w1), np.max(w2)])
plt.plot([min_weight, max_weight], [min_weight, max_weight], color = 'k', linestyle = '--', linewidth = 2)
```

# Select people wigh the highest w1

```python
order_w1 = np.argsort(w1) # Sort the w1 values from low to high
w1_high = order_w1[-n_high:] # Select the n_high largest values from the sorted list of w1

# Plot w1 and w2 as before, with the highest w1 values highlighted in red
plt.figure(figsize = [4, 4])
plt.scatter(w1, w2, color = 'w', marker = 'o', edgecolor = 'k', s = 20, alpha = 0.5)
plt.scatter([w1[i] for i in w1_high], [w2[i] for i in w1_high], color = 'w', marker = 'o', edgecolor = 'r',
           s = 20)
plt.xlabel('w1')
plt.ylabel('w2')
plt.plot([min_weight, max_weight], [min_weight, max_weight], color = 'k', linestyle = '--', linewidth = 2)
```

Zoom in on the red dots:

```python
plt.figure(figsize = [4, 4])
plt.scatter([w1[i] for i in w1_high], [w2[i] for i in w1_high], color = 'w', marker = 'o', edgecolor = 'r',
           s = 50, linewidth = 1.8)
plt.xlabel('w1')
plt.ylabel('w2')
plt.plot([w1[order_w1[-n_high]] - 2*std_weight, w1[order_w1[-1]] + 2*std_weight], 
         [w1[order_w1[-n_high]] - std_weight, w1[order_w1[-1]] + std_weight], color = 'k', 
         linestyle = '--', 
         label = 'w1 = w2 line')
plt.xlim([w1[order_w1[-n_high]] - 4*std_weight, w1[order_w1[-1]] + 4*std_weight])
plt.ylim([w1[order_w1[-n_high]] - 4*std_weight, w1[order_w1[-1]] + 4*std_weight])

print('On average, the difference between w1 and w2 for these people with the highest w1 values is {0} kg.'.format(
    np.round(np.mean([w2[i] - w1[i] for i in w1_high]), 2)))
```

# Show that there's no causal effect, by demonstrating the same thing happens when you select people with the highest w2 values

```python
order_w2 = np.argsort(w2) # Sort the w2 values from low to high
w2_high = order_w2[-n_high:] # Select the n_high largest values from the sorted list of w2

# Plot w1 and w2 as before, with the highest w2 values highlighted in red
plt.figure(figsize = [4, 4])
plt.scatter(w1, w2, color = 'w', marker = 'o', edgecolor = 'k', s = 20, alpha = 0.5)
plt.scatter([w1[i] for i in w2_high], [w2[i] for i in w2_high], color = 'w', marker = 'o', edgecolor = 'r',
           s = 20)
plt.xlabel('w1')
plt.ylabel('w2')
plt.plot([min_weight, max_weight], [min_weight, max_weight], color = 'k', linestyle = '--', linewidth = 2)
```

Zoom in on the red dots:

```python
plt.figure(figsize = [4, 4])
plt.scatter([w1[i] for i in w2_high], [w2[i] for i in w2_high], color = 'w', marker = 'o', edgecolor = 'r',
           s = 50, linewidth = 1.8)
plt.xlabel('w1')
plt.ylabel('w2')
plt.plot([w2[order_w2[-n_high]] - 2*std_weight, w2[order_w1[-1]] + 2*std_weight], 
         [w2[order_w2[-n_high]] - std_weight, w2[order_w1[-1]] + std_weight], color = 'k', 
         linestyle = '--', 
         label = 'w1 = w2 line')
plt.xlim([w2[order_w2[-n_high]] - 4*std_weight, w2[order_w2[-1]] + 4*std_weight])
plt.ylim([w2[order_w2[-n_high]] - 4*std_weight, w2[order_w2[-1]] + 4*std_weight])

print('On average, the difference between w1 and w2 for these people with the highest w2 values is {0} kg.'.format(
    np.round(np.mean([w2[i] - w1[i] for i in w2_high]), 2)))
```
