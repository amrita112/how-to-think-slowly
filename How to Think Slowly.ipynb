# Regression to the Mean

## Weight on two different days

### Parameters

```python
n = 10000 # Number of people
n_high = 100 # Number selected for demonstrating regression

min_weight = 40
max_weight = 100
pop_mean_weight = 70 # Population mean weight
pop_std_weight = 10 # Standard deviation of population weight (used when weight_dist = 'normal')
weight_dist = 'normal'
std_weight = 2 # Standard deviation for individual's measurements
```

### Generate measurements

```python
if weight_dist == 'uniform':
    mean_weight = np.random.rand(n)*(max_weight - min_weight) + min_weight
elif weight_dist == 'normal':
    mean_weight = np.random.normal(loc = pop_mean_weight, scale = pop_std_weight, size = n)

w1 = np.random.normal(loc = mean_weight, scale = std_weight)
w2 = np.random.normal(loc = mean_weight, scale = std_weight)

order_mean = np.argsort(mean_weight)
emp_min = np.min([np.min(w1), np.min(w2)])
emp_max = np.max([np.max(w1), np.max(w2)])

plt.figure()
plt.plot([w1[i] for i in order_mean], color = 'r', label = 'w1')
plt.plot([w2[i] for i in order_mean], color = 'b', label = 'w2')
plt.plot(np.sort(mean_weight), color = 'k', label = 'mean')
plt.legend()

plt.figure()
plt.hist(mean_weight)
plt.xlabel('Weight')
plt.ylabel('Number of people')
print('Mean w1 = {0}'.format(np.mean(w1)))
print('Mean w2 = {0}'.format(np.mean(w2)))
print('Mean difference = {0}'.format(np.mean(w1 - w2)))
```

```python
plt.figure(figsize = [4, 4])
plt.scatter(w1, w2, color = 'w', marker = 'o', edgecolor = 'k', s = 20, alpha = 0.5)
#plt.xticks([])
#plt.yticks([])
#plt.xlabel('Weight - measurement 1')
#plt.ylabel('Weight - measurement 2')
plt.xlabel('w1')
plt.ylabel('w2')
plt.plot([emp_min, emp_max], [emp_min, emp_max], color = 'k', linestyle = '--', 
         label = 'w1 = w2 line')
#plt.legend()
plt.savefig('{0}{1}w1_vs_w2.png'.format(fig_folder, sep), bbox_inches = 'tight')
```

### Select Highest w1

```python
order_w1 = np.argsort(w1)
w1_high = order_w1[-n_high:]

plt.figure(figsize = [4, 4])
plt.scatter(w1, w2, color = 'w', marker = 'o', edgecolor = 'k', s = 20, alpha = 0.5)
plt.scatter([w1[i] for i in w1_high], [w2[i] for i in w1_high], color = 'w', marker = 'o', edgecolor = 'r',
           s = 20)
#plt.xticks([])
#plt.yticks([])
#plt.xlabel('Weight - measurement 1')
#plt.ylabel('Weight - measurement 2')
plt.xlabel('w1')
plt.ylabel('w2')
plt.plot([emp_min, emp_max], [emp_min, emp_max], color = 'k', linestyle = '--', 
         label = 'w1 = w2 line')
#plt.legend()
plt.savefig('{0}{1}w1_vs_w2_100_highest_w1.png'.format(fig_folder, sep), bbox_inches = 'tight')
```

```python
plt.figure(figsize = [4, 4])
#plt.scatter(w1, w2, color = 'w', marker = 'o', edgecolor = 'k', s = 20)
plt.scatter([w1[i] for i in w1_high], [w2[i] for i in w1_high], color = 'w', marker = 'o', edgecolor = 'r',
           s = 50, linewidth = 1.8)
#plt.xticks([])
#plt.yticks([])
#plt.xlabel('Weight - measurement 1')
#plt.ylabel('Weight - measurement 2')
plt.xlabel('w1')
plt.ylabel('w2')
plt.plot([w1[order_w1[-n_high]] - 2*std_weight, w1[order_w1[-1]] + 2*std_weight], 
         [w1[order_w1[-n_high]] - std_weight, w1[order_w1[-1]] + std_weight], color = 'k', 
         linestyle = '--', 
         label = 'w1 = w2 line')
#plt.legend()
plt.xlim([w1[order_w1[-n_high]] - 4*std_weight, w1[order_w1[-1]] + 4*std_weight])
plt.ylim([w1[order_w1[-n_high]] - 4*std_weight, w1[order_w1[-1]] + 4*std_weight])

print('Mean weight loss = {0}'.format(np.mean([w2[i] - w1[i] for i in w1_high])))
plt.savefig('{0}{1}zoom_100_highest_w1.png'.format(fig_folder, sep), bbox_inches = 'tight')
```

## Bimodal distribution

### Generate measurements

```python
fem_weights = np.random.normal(loc = 40, scale = 10, size = 5000)
male_weights = np.random.normal(loc = 100, scale = 10, size = 5000)
mean_weight = np.concatenate([fem_weights, male_weights])

w1 = np.random.normal(loc = mean_weight, scale = std_weight)
w2 = np.random.normal(loc = mean_weight, scale = std_weight)

order_mean = np.argsort(mean_weight)
emp_min = np.min([np.min(w1), np.min(w2)])
emp_max = np.max([np.max(w1), np.max(w2)])

plt.figure()
plt.plot([w1[i] for i in order_mean], color = 'r', label = 'w1')
plt.plot([w2[i] for i in order_mean], color = 'b', label = 'w2')
plt.plot(np.sort(mean_weight), color = 'k', label = 'mean')
plt.legend()

plt.figure()
plt.hist(mean_weight)
plt.xlabel('Weight')
plt.ylabel('Number of people')
print('Mean w1 = {0}'.format(np.mean(w1)))
print('Mean w2 = {0}'.format(np.mean(w2)))
print('Mean difference = {0}'.format(np.mean(w1 - w2)))
```

### Select highest w1

```python
order_w1 = np.argsort(w1)
w1_high = order_w1[-n_high:]

plt.figure(figsize = [4, 4])
plt.scatter(w1, w2, color = 'w', marker = 'o', edgecolor = 'k', s = 20, alpha = 0.2)
plt.scatter([w1[i] for i in w1_high], [w2[i] for i in w1_high], color = 'w', marker = 'o', edgecolor = 'r',
           s = 20)
#plt.xticks([])
#plt.yticks([])
#plt.xlabel('Weight - measurement 1')
#plt.ylabel('Weight - measurement 2')
plt.xlabel('w1')
plt.ylabel('w2')
plt.plot([emp_min, emp_max], [emp_min, emp_max], color = 'k', linestyle = '--', 
         label = 'w1 = w2 line')
#plt.legend()
plt.savefig('{0}{1}w1_vs_w2_100_highest_w1.png'.format(fig_folder, sep), bbox_inches = 'tight')
```
